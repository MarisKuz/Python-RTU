```python

 
import requests
import json
import pandas as pd
import time
url = "http://www.recipepuppy.com/api/?q=ingr"
 
def get_json_results(url):
    response = requests.get(url)
    if response.status_code != 200:
        return []
    return response.json().get('results')  
 
def get_single_page(url, to_excel = False):
    df = pd.DataFrame(get_json_results(url))
    df = df[['title', 'ingredients']]
#     df.head()  
#     df.tail()
    if to_excel:
        df.to_excel(f"{ingr}.xlsx")
    return df

def main(ingr):
    dfs = []
    for p in range(1,11):
        url = f"http://www.recipepuppy.com/api/?q={ingr}&p={p}"
        print(f"Retrieving: {url}")
        time.sleep(0.9)
        df = get_single_page(url)
        dfs.append(df)
        
    df = pd.concat(dfs)
    return df
```

```python
ingr = input('Write down one ingredient like potato, selery, onions etc...:  ')
df = main(ingr)
df
```
