```python
import random
from random import randint as rnd # random integer N, A ≤ N ≤ B. shorted
from collections import Counter
from collections import OrderedDict # to remember the order entries data added
import matplotlib.pyplot as plt 
```

```python
def dice_throws(dices_cnt=6, throws_ln=100000):   # setup parameters of throws and throws count
    return [sum(rnd(1,6) for throw in range(dices_cnt)) for throwing in range(throws_ln)] # dice throw from 1 to 6
```

```python
# # Uzd No.1

throw_lst = dice_throws(throws_ln = 20_000)         #creating list to sort & order
freq = Counter(throw_lst)   # making dictionary
 
sorting = OrderedDict(sorted(freq.items(), key=lambda t: t[0])) # remembers the order entries data added to dict and sort it in sequence 6 to 36 
                                                              
#  # Print/plot a graph using pyplot
plt.figure(figsize=(15,10))  #plot a blank graph with 
plt.bar(range(len(sorting)), list(sorting.values()), align='center', color='purple')
plt.xticks(range(len(sorting)), list(sorting.keys()))
plt.grid()  #dispalys a grid # plt.ion() if live changes will be required then could be used
plt.show()  #display all open figures
```
